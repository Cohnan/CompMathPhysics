```python
from scipy.interpolate import lagrange
from sympy import *

init_printing(use_unicode = True)
```

```python
h, t, tn, fn, fn1, fn2, fn3 = symbols('h t t_n f_{n} f_{n-1} f_{n-2} f_{n-3}')

tn1 = tn - h
tn2 = tn1 - h
tn3 = tn2 - h
```

### Orden 2

```python
simplify(
    integrate(
        fn * (t - tn1)/(tn - tn1) + fn1 * (t - tn)/(tn1 - tn),
        (t, tn, tn + h))
    )
```

### Orden 3

```python
simplify(
    integrate(
        fn  * ((t - tn1) * (t - tn2))/((tn - tn1) * (tn - tn2)) + 
        fn1 * ((t - tn) * (t - tn2))/((tn1 - tn) * (tn1 - tn2)) +
        fn2 * ((t - tn) * (t - tn1))/((tn2 - tn) * (tn2 - tn1)),
        (t, tn, tn + h))
    )
```

### Orden 4

```python
simplify(
    integrate(
        fn  * ((t - tn1) * (t - tn2) * (t - tn3))/((tn - tn1) * (tn - tn2) * (tn - tn3)) + 
        fn1 * ((t - tn) * (t - tn2) * (t - tn3))/((tn1 - tn) * (tn1 - tn2) * (tn1 - tn3)) +
        fn2 * ((t - tn) * (t - tn1) * (t - tn3))/((tn2 - tn) * (tn2 - tn1) * (tn2 - tn3)) +
        fn3 * ((t - tn) * (t - tn1) * (t - tn2))/((tn3 - tn) * (tn3 - tn1) * (tn3 - tn2)),
        (t, tn, tn + h))
    )
```

## 2.

```python
from scipy.integrate import odeint
```

```python
omega0 = 1.
zeta = 1.

def derivada_vec(Y, t):
    x = Y[0]
    v = Y[1]
    
    return array([1./(2*zeta*omega0) * (-omega0**2 * x + 2*zeta*omega0*v + omega0**2 *x),
                 - 2*zeta*omega0*v - omega0**2 * x])
```

```python
ts = np.linspace(0., 8*np.pi, 10000)
t0 = array([1., 0.])

fig = plt.figure("Oscilador amortiguado", figsize = (24, 8))

# Subamortiguado

zeta = 0.1

ax1 = fig.add_subplot(131)

Ys =  odeint(derivada_vec, t0, ts, printmessg=True)

xs = Ys[:, 0]
vs = Ys[:, 1]

ax1.plot(ts, xs)

# Amortigamiento crítico

zeta = 1.

ax2 = fig.add_subplot(132)

Ys =  odeint(derivada_vec, t0, ts, printmessg=True)

xs = Ys[:, 0]
vs = Ys[:, 1]

ax2.plot(ts, xs)

# Sobreamortiguado

zeta = 10.

ax3 = fig.add_subplot(133)

Ys =  odeint(derivada_vec, t0, ts, printmessg=True)

xs = Ys[:, 0]
vs = Ys[:, 1]

ax3.plot(ts, xs)
```
