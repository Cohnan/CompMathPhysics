```python
%pylab inline
import scipy.integrate as integrate
```

```python
L = 3.

def a(n, x):
    return 1/L * integrate.quad(lambda x: original1(x) * np.cos(n * np.pi* x / L), -L, L)[0]
    
def b(n, x):
    return 1/L * integrate.quad(lambda x: original1(x) * np.sin(n * np.pi* x / L), -L, L)[0]

def funcion_parcial(x, n):
    suma = 0.5 * a(0, x)
    
    for i in range(1, 1 + n):
        suma += a(i, x) * cos(i * np.pi* x / L) + b(i, x) * sin(i * np.pi* x / L)
    
    return suma

def original1(x):
    xe = x % L
    
    if xe < 1:
        return xe
    
    return 1 - 0.5 * (xe - 1)
```

```python
x = np.linspace(0, 3, 100)
yo = np.array([original1(el) for el in x])

fig = plt.figure(1, figsize=(10, 5))
plt.plot(x, yo)

for n in range(2, 8):
    y = np.array([funcion_parcial(el, n) for el in x])
    plt.plot(x, y)
```
