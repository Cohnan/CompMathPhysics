```python
theta = linspace(0, 2 * np.pi, 100)
```

```python
for i in range(1, 6):
    for j in range(1,6):
        plt.subplot(5, 5, 5*(i-1) + j)
        plot(np.cos(i*theta), np.sin(j*theta))

plt.show()
```
