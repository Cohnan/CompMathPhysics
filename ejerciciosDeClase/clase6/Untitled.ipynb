```python
theta = linspace(0, 2 * np.pi, 100)

fig = figure('Lissajous', figsize = (8,8))

for i in range(1, 6):
    for j in range(1,6):
        ax = fig.add_subplot(5, 5, 5*(i-1) + j)
        ax.plot(np.cos(i*theta), np.sin(j*theta))
        ax.axis('off')
savefig('liss.png')
savefig('liss.pdf')
show()
```
