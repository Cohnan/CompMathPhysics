Se estudian diferentes patrones de difracción producidos por una apertura, en el régimen de Fraunhofer(es decir, cuando el plano de incidencia se encuentra en el campo lejano). 

Para ésto se utilizan diferentes *pupil functions*, que son las funciones coeficientes de las integrales de Fourier que determinan la amplitud compleja de la luz en cada punto del plano de incidencia. Estas *pupil functions* son simplemente funciones en la porción de $\mathbb{R}^2$ que representa la placa de la rendija que toman un valor constante en la apertura y 0 en caso contrario.

Aquí utilizaremos de una placa cuadrada de $1u$ de lado, representada por un array cuadrado de lado $2^{10}$. El patrón se hará de acuerdo a la irradiancia de la luz, la cual es proporcional al cuadrado de la amplitud compleja (Las constantes serán ignoradas debido a que estas son de por sí para normalización, que es llevada a cabo aquí al hacer la graficación con la función `imshow`). Además, aproximaremos las integrales de Fourier con integrales de Fourier discretas para obtener el patrón de interferencia.

A. Creación de rendija cuadrada de lado (en array) de $2^5$

```python
largoG = 2**10 # Largo de la placa
largoP = 2**5  # Largo de la apertura

# Incialización de la rendija. 
# Primero en 0's y luego cambio del centro a un cuadrado abierto
rendija1 = np.zeros((largoG, largoG), dtype = float32)
rendija1[(largoG - largoP)/2 : (largoG + largoP)/2, (largoG - largoP)/2 : (largoG + largoP)/2] = 1
```

B. Cálculo de patrón de interferencia. Se grafica el logaritmo de la intensidad a cambio.

```python
# Cálculo de amplitud compleja
rendija1_dft = fft2(rendija1)
rendija1_shift = fftshift(rendija1_dft) # Acomodación de los datos para su graficación

# Lado (en cuadros de array) de la región a mostrar del plano de incidencia
lado_most1 = 200

# Cálculo del (logaritmo) de la intensidad en la región a mostrar
mostrar1 = np.log(
    np.absolute(
        rendija1_shift[(largoG - lado_most1)/2 : (largoG + lado_most1)/2, (largoG - lado_most1)/2 : (largoG + lado_most1)/2]
    )**2 + 1) #Sin el 1, la gráfica es idéntica a la del enunciado

# Graficación de la rendija utilizada y el patrón resultante en un cuadro
# centrado de lado indicado
fig = plt.figure('Rendija Cuadrada', figsize = (10, 5))

ax1 = fig.add_subplot(121)
ax1.imshow(rendija1, cmap = 'gray')
ax1.set_title(u"Rendija Cuadrada")

ax2 = fig.add_subplot(122)
ax2.imshow(mostrar1)
ax2.set_title(u"Patrón de Interferencia\n(log resaltado)")

plt.show( )
```

C. Patrón de interferencia resultante de rendija con doble apertura

```python
# Dimensión de las aperturas
largoP = 200

grosor = 20
altura = 200

# Coordenadas de lugares de inicio de las aperturas
inicio_1 = 459
inicio_2 = 564

# Lado del cuadro a mostrar del plano de incidencia
lado_most2 = 128

# Creación de la placa con las aperturas
rendija2 = np.zeros((largoG, largoG), dtype = float32)
rendija2[(largoG - largoP)/2 : (largoG + largoP)/2, inicio_1 : inicio_1 + grosor] = 1
rendija2[(largoG - largoP)/2 : (largoG + largoP)/2, inicio_2 : inicio_2 + grosor] = 1

# Cálculo de las amplitudes complejas producidas por la rendija
rendija2_dft = fft2(rendija2)
rendija2_shift = fftshift(rendija2_dft)

# Cálculo de intensidades en región a mostrar
mostrar2 = np.absolute(
        rendija2_shift[(largoG - lado_most2)/2 : (largoG + lado_most2)/2, (largoG - lado_most2)/2 : (largoG + lado_most2)/2]
    )**2

# Graficación de rendija y del patrón en la región indicada (cuadro centrado)
fig = plt.figure('Rendija Doble', figsize = (10, 5))

ax1 = fig.add_subplot(121)
ax1.imshow(rendija2, cmap = 'gray')
ax1.set_title(u"Rendija Doble")

ax2 = fig.add_subplot(122)
ax2.imshow(mostrar2)
ax2.set_title(u"Patrón de Interferencia")

plt.show( )
```

D. Apertura circular de radio $1/100 u$, con gráfica de $\frac{intensidad}{intensidad\ máxima}$ en la línea central.

```python
largoG = 2**10
radioP = 0.01*largoG

# Lado de la región a mostrar
lado_most3 = largoG

# Creacioń de la rendija
rendija3 = np.zeros((largoG, largoG), dtype = float32)

for i in range(largoG):
    # Marca como parte de la apertura aquellos puntos que pertenecen a un
    # círculo centrado basándose en las coordenadas dadas por los índices del
    # array.
    
    for j in range(largoG):
        if (i + 0.5 - largoG/2)**2 + (j + 0.5 - largoG/2)**2 <= radioP**2:
            rendija3[i, j] = 1

# Cálculo de amplitudes complejas
rendija3_dft = fft2(rendija3)
rendija3_shift = fftshift(rendija3_dft)

# Cálculo del (logaritmo) de la intensidad en la región a mostrar
mostrar3 = np.log(
    np.absolute(
        rendija3_shift[(largoG - lado_most3)/2 : (largoG + lado_most3)/2, (largoG - lado_most3)/2 : (largoG + lado_most3)/2]
    )**2 + 1)

# Graficación de rendija y patrón de interferencia
fig = plt.figure('Rendija Circular', figsize = (10, 10))

ax1 = fig.add_subplot(223)
ax1.imshow(rendija3, cmap = 'gray', interpolation = 'Nearest')
ax1.set_title(u"Rendija Doble")

ax2 = fig.add_subplot(224)
ax2.imshow(mostrar3)
ax2.set_title(u"Patrón de Interferencia")

ax3 = fig.add_subplot(211)

# Gráfica de intensidad / intensidad máxima en la línea central de la región a mostrar
x = 2*np.pi*radioP/1024.*np.linspace(-512,512,1024)
y = np.exp(mostrar3[largoG / 2,:])/np.exp(mostrar3[largoG / 2, largoG / 2])

# Regresión de la amplitud compleja generada, en el plano de incidencia
from scipy.special import jv # Función de Bessel de la primera clase

def funcion(x, c):
    # Función teórica que determina la amplitud compleja de la luz en el plano de incidencia
    
    return (2*jv(1, c*x) / (c*x))**2

pars, cov = curve_fit(funcion, x, y)
c = pars[0]


ax3.set_xlim([0, 10])
ax3.plot(x, y, 'bo', ms = 2, label = 'Datos Experimentales')
ax3.plot(x, funcion(x, c), 'r-', label = 'Curve Fit')
ax3.set_ylabel('$I/I_0$')
ax3.set_xlabel('$2\pi Rj/L$')
ax3.legend( )

plt.show( )
```

E. Tabla de máximos y mínimos entre $0$ y $12$ de la abscisa de la gráfica anterior.

```python
from scipy.signal import argrelextrema as relex  # Función para calcular los valores extremos

# 
new_arr_x = array([e for e in x if (e < 12 and e >=0)])
new_arr_y = array([funcion(e, 1.00371387) for e in new_arr_x])

ind1 = relex(new_arr_y, np.greater)
xs1 = [new_arr_x[i] for i in ind1]

print ind1
print xs1
print [funcion(e, 1.00371387) for e in xs1]

ind2 = relex(new_arr_y, np.less)
xs2 = [new_arr_x[i] for i in ind2]

print ind2
print xs2
print [funcion(e, 1.00371387) for e in xs2]
```
